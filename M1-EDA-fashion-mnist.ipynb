```python
import pandas as pd
import numpy as np
import sweetviz as sv
from tensorflow.keras.datasets import fashion_mnist
from ydata_profiling import ProfileReport

# Load Fashion MNIST dataset 
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

# ✅ Combine training and test data
x_full = np.concatenate((x_train, x_test), axis=0)
y_full = np.concatenate((y_train, y_test), axis=0)

# Convert images to tabular format (flattening 28x28 images to 1D)
x_full_flat = x_full.reshape(x_full.shape[0], -1)

# Convert to DataFrame
df_full = pd.DataFrame(x_full_flat)
df_full['label'] = y_full  # Add target column

df_sample = df_full  # Use full data

profile = ProfileReport(
    df_sample, title="Fashion MNIST EDA Report", explorative=True, minimal=True
)

# Save the report as an HTML file
profile.to_file("fashion_mnist_full_eda.html")
sweetviz_report = sv.analyze(df_sample, pairwise_analysis="off")  
sweetviz_report.show_html("fashion_mnist_sweetviz.html")

print(" EDA Reports Generated:")
print("- Pandas Profiling: fashion_mnist_pandas_profiling.html")
print("- Sweetviz: fashion_mnist_sweetviz.html")
```

Observations -
* The ydata profiling ans sweetviz , both give almost the same kind of report . These reports focuses on the correlations of the features.
* The report interprets that there are no missing values within our used dataset. When the datset changes in future these profiling reports become useful to analyse the columns and features and their relations in our data.
